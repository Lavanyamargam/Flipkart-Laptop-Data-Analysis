### RAM Size Distribution

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Define standard RAM sizes
standard_ram = [4, 8, 16, 32]

# Get counts for only standard RAM sizes
ram_counts = df["RAM_GB"].value_counts().loc[standard_ram]

# Create barplot
plt.figure(figsize=(8,5))
ax = sns.barplot(x=ram_counts.index, y=ram_counts.values, palette="viridis")

# Add count labels on top of bars
for i, val in enumerate(ram_counts.values):
    ax.text(i, val, str(val), ha='center', va='bottom')

# Titles and labels
plt.title("Distribution of RAM Sizes")
plt.xlabel("RAM Size (GB)")
plt.ylabel("Count")
plt.show()
```

- This shows how many laptops have 4GB, 8GB, 16GB, etc.
-  8GB RAM dominates, followed by 16GB. 4GB is still present in budget laptops.


```python
import matplotlib.pyplot as plt
import seaborn as sns

# Map near-256 values to 256 GB
df["Storage_GB_clean"] = df["Storage_GB"].replace({260:256, 264:256, 268:256, 272:256})

# Filter out rare values (count <= 2)
storage_counts = df["Storage_GB_clean"].value_counts()
filtered_storage = storage_counts[storage_counts > 2]

# Plot
plt.figure(figsize=(10,6))
ax = sns.barplot(x=filtered_storage.index, y=filtered_storage.values, palette="mako")

# Add count labels
for i, val in enumerate(filtered_storage.values):
    ax.text(i, val, str(val), ha='center', va='bottom')

plt.title("Distribution of Storage Capacities")
plt.xlabel("Storage Size (GB)")
plt.ylabel("Count")
plt.show()
```

- 512GB SSD is most common, followed by 256GB. 1TB appears in premium laptops.


```python
# Histogram of Display_Size_Inch
plt.figure(figsize=(8,5))
sns.histplot(df["Display_Size_Inch"], bins=15, kde=False, color="orange")

plt.title("Distribution of Display Sizes")
plt.xlabel("Display Size (inches)")
plt.ylabel("Count")
plt.show()
```

- 15.6 inch is the most popular size, followed by 14 inch. Smaller screens (11.6 inch) are mainly Chromebooks.

```python
# Good feedback (Rating >= 4.3)
good_feedback = df[df["Rating"] >= 4.3]

# Bad feedback (Rating <= 3.9)
bad_feedback = df[df["Rating"] <= 3.9]

plt.figure(figsize=(8,5))
sns.histplot(good_feedback["Rating"], bins=10, color="green", label="Good Feedback", kde=True)
sns.histplot(bad_feedback["Rating"], bins=10, color="red", label="Bad Feedback", kde=True)
plt.title("Distribution of Good vs Bad Feedback")
plt.xlabel("Rating")
plt.ylabel("Count")
plt.legend()
plt.show()
```

### Feedback Distribution
- Good feedback (≥4.3) is common in mid-range and premium laptops.
- Negative feedback (≤3.9) clusters around budget laptops with Celeron processors.
- Customer satisfaction is strongly tied to specs and brand reputation.

```python
segments = {
    "Students": df[(df["RAM_GB"]==8) & (df["Price"]<=40000)],
    "Professionals": df[(df["RAM_GB"]==16) & (df["Price"].between(50000,90000))],
    "Gamers": df[(df["RAM_GB"]>=16) & (df["Price"]>=70000)],
    "Common Users": df[(df["RAM_GB"]==8) & (df["Price"]<=30000)]
}

avg_prices = {seg: data["Price"].mean() for seg, data in segments.items()}

plt.figure(figsize=(8,5))
ax = sns.barplot(x=list(avg_prices.keys()), y=list(avg_prices.values()), palette="Set2")

# Add labels
for p in ax.patches:
    ax.annotate(f'{p.get_height():.0f}',
                (p.get_x() + p.get_width()/2., p.get_height()),
                ha='center', va='bottom')

plt.title("Average Price by User Segment")
plt.xlabel("Segment")
plt.ylabel("Average Price (INR)")
plt.show()
```

### Average Price by User Segment
- Students: ~₹35k (affordable laptops with 8GB RAM).
- Professionals: ~₹65k (balanced productivity machines).
- Gamers: ~₹85k–₹95k (high-performance with GPUs).
- Common Users: ~₹28k (basic browsing/office laptops).

```python
tiered = df.groupby("Price_Tier")["Price"].mean().reset_index()

plt.figure(figsize=(6,4))
ax = sns.barplot(data=tiered, x="Price_Tier", y="Price", palette="coolwarm")

# Add labels
for p in ax.patches:
    ax.annotate(f'{p.get_height():.0f}',
                (p.get_x() + p.get_width()/2., p.get_height()),
                ha='center', va='bottom')

plt.title("Average Price by Tier")
plt.xlabel("Tier")
plt.ylabel("Average Price (INR)")
plt.show()
```

### Tiered Pricing Strategy
- Budget laptops average ~₹30k.
- Mid-range laptops average ~₹55k–₹65k.
- Premium laptops average ~₹90k–₹1 lakh.
- Sellers can set competitive prices by aligning with these tiers.

```python
cpu_price = df.groupby("CPU_Brand")["Price"].mean().reset_index()

plt.figure(figsize=(8,5))
ax = sns.barplot(data=cpu_price, x="CPU_Brand", y="Price", palette="muted")

# Add labels
for p in ax.patches:
    ax.annotate(f'{p.get_height():.0f}',
                (p.get_x() + p.get_width()/2., p.get_height()),
                ha='center', va='bottom')

plt.title("Average Price by CPU Brand")
plt.xlabel("CPU Brand")
plt.ylabel("Average Price (INR)")
plt.xticks(rotation=45)
plt.show()
```

### Feature-Based Pricing (CPU Brand)

# 💡 Competitive Pricing Strategies for Sellers

### 1. **Tiered Pricing Strategy**
- Offer laptops in **Budget (₹20k–₹35k)**, **Mid-range (₹40k–₹70k)**, and **Premium (₹80k+)** tiers.  
- Example: Acer Aspire 3 (Budget), ASUS Vivobook 16X (Mid-range), Apple MacBook Pro (Premium).

---

### 2. **Bundle Pricing**
- Add value by bundling laptops with accessories (bag, mouse, extended warranty).  
- Example: HP Pavilion 14 + wireless mouse + 1-year warranty at a slightly higher price.

---

### 3. **Seasonal Discounts**
- Offer festive or back-to-school discounts to attract students and professionals.  
- Example: ASUS Vivobook Go 15 at ₹29,990 during exam season.

---

### 4. **Dynamic Pricing**
- Adjust prices based on demand and competitor listings.  
- Example: Lower Acer Aspire 3 price when ASUS Vivobook Go 15 is trending.

---

### 5. **Feature-Based Differentiation**
- Price laptops based on unique features (touchscreen, gaming GPU, ultrathin design).  
- Example: Lenovo Legion 5 priced higher due to RTX GPU compared to a standard Lenovo IdeaPad.
